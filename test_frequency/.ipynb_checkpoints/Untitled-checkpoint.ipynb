```python
import os
```

```python
nouns = []
with open('5letter_nouns.txt', 'r') as f:
    nouns = f.read().split('\n')
nouns_counter = {x:0 for x in nouns}
```

```python
for idx, file in enumerate(os.listdir('books')):
    with open(f'books/{file}', 'r') as f:
        print(file)
        book_text = f.read()
        for noun in nouns:
            nouns_counter[noun] += book_text.count(noun)
            nouns_counter[noun] += book_text.count(noun.upper())
            nouns_counter[noun] += book_text.count(noun.capitalize())
```

```python
nouns_counter
```

```python
import json
```

```python
with open('noun_frequency.json', 'w', encoding="UTF-8") as f:
    json.dump(nouns_counter, f, ensure_ascii=False)
```
